```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load the dataset
file_path = '/Users/juliuswalkenhorst/masterthesis/datapreparation/electionmastermerged.csv'
df = pd.read_csv(file_path)
```

```python
# Parse time columns to datetime
df['time'] = pd.to_datetime(df['time'])
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])

# Create a new column for time relative to event
df['minutes_from_event'] = (df['time'] - df['event_timestamp']).dt.total_seconds() / 60

# Filter data to only include +/- 24 hours (i.e., +/- 1440 minutes)
df = df[(df['minutes_from_event'] >= -1440) & (df['minutes_from_event'] <= 1440)]

# Group by label (event)
labels = df['label'].unique()
```

```python
# Create a plot for each event
for label in labels:
    df_event = df[df['label'] == label]

    # Create a figure
    plt.figure(figsize=(12, 6))

    # Plot each sharesType separately
    for sharesType in df_event['sharesType'].unique():
        df_shares = df_event[df_event['sharesType'] == sharesType]
        
        # Sort by time to make the line plot correct
        df_shares = df_shares.sort_values('time')
        
        plt.plot(df_shares['time'], df_shares['probabilities'], label=sharesType)

    # Plot vertical line at T0
    t0 = df_event['event_timestamp'].iloc[0]
    plt.axvline(t0, color='red', linestyle='--', label='Event T₀')

    # Format plot
    plt.title(f'Event {label}: Price (Probabilities) vs. Time')
    plt.xlabel('Time')
    plt.ylabel('Probability')
    plt.legend()
    plt.grid(True)

    # Format x-axis for better readability
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gcf().autofmt_xdate()

    # Show the plot
    plt.show()
```
