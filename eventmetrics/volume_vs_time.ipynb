```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Load the dataset
file_path = '/Users/juliuswalkenhorst/masterthesis/datapreparation/electionmastermerged.csv'
df = pd.read_csv(file_path)
```

```python
# Parse time columns to datetime
df['time'] = pd.to_datetime(df['time'])
df['event_timestamp'] = pd.to_datetime(df['event_timestamp'])
```

```python
# Define the time window (in minutes)
time_window = 1140  # 60 minutes for 1 hour

# Filter data to only include +/- time_window
df = df[(df['minutes_from_event'] >= -time_window) & (df['minutes_from_event'] <= time_window)]

# Group by label (event)
labels = df['label'].unique()

# Create a plot for each event
for label in labels:
    df_event = df[df['label'] == label]

    # Create a figure
    plt.figure(figsize=(12, 6))

    # Determine bin size based on time window
    if time_window == 60:
        bin_size = '1min'
    else:
        bin_size = '5min'

    # Plot each sharesType separately
    for sharesType in df_event['sharesType'].unique():
        df_shares = df_event[df_event['sharesType'] == sharesType]
        df_shares = df_shares.set_index('time')

        # Resample buyerPaid only
        df_resampled = df_shares['buyerPaid'].resample(bin_size).sum().reset_index()

        plt.bar(df_resampled['time'], df_resampled['buyerPaid'], width=0.003, label=sharesType, align='center')

    # Plot vertical line at T0
    t0 = df_event['event_timestamp'].iloc[0]
    plt.axvline(t0, color='red', linestyle='--', label='Event T')

    # Format plot
    plt.title(f'Event {label}: Volume (BuyerPaid) vs. Time')
    plt.xlabel('Time')
    plt.ylabel('Volume (BuyerPaid)')
    plt.legend()
    plt.grid(True)

    # Format x-axis for better readability
    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d %H:%M'))
    plt.gcf().autofmt_xdate()

    # Show the plot
    plt.show()
```
